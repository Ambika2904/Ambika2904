```python
import random
import periodictable

def first_20_elements():
    return [periodictable.elements[i] for i in range(1, 21)]  #using range to use only first 20 elements from periodic table

def display_element(element):
    print(f"\nElement: {element.name} (Symbol: {element.symbol})") #for game we will give element name and symbol

def start():
    elements = first_20_elements() #only between 20 elements
    players = []

    # Getting number of players with names
    num_players = int(input("Enter the number of players: ")) #to know how many member will play
    for i in range(num_players):
        name = input(f"Enter the name of player {i + 1}: ") #adding their names
        players.append(name)

    rounds_played = 0 #computer starts the numbers from 0 means 1st round

    while True:
        element = random.choice(elements) #will show random element name with symbol and player will enter atomic number
        display_element(element)

        for player in players:
            guess = int(input(f"{player}, enter your guess for the atomic number: ")) #this is to guess the atomic number
            if guess == element.number:
                print(f"Congratulations, {player}! You guessed correctly!")
                break
        else:
            print("No one guessed correctly this round.")

        print(f"The correct atomic number was: {element.number}.") #if no one guess right answer then will show the correct answer
        print(f"The element was: {element.name} (Symbol: {element.symbol})")

        rounds_played += 1 #to contiune the rounds

        if rounds_played >= 10: #only till 10 round and then will ask user if they want to continue
            if input("You've played 10 rounds. Do you want to continue? (yes/no): ").strip().lower() != 'yes':
                print("Thanks for playing!")
                break

# Directly call start()
start()
```
